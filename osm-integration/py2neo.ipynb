```python
from py2neo import Graph
from py2neo.bulk import create_nodes, merge_relationships
from utils import GraphNode, GraphEdge, haversine
from itertools import islice
```

```python
import os
from dotenv import load_dotenv
load_dotenv()

g = Graph(os.getenv('CON_STRING'), auth=(os.getenv('CON_USER'),os.getenv('CON_PASS')))
```

```python
import osmium
class FileHandler(osmium.SimpleHandler):
    def __init__(self, neoHandler):
        super(FileHandler, self).__init__()
        self.neoHandler = neoHandler

    def node(self, n):
        node = GraphNode(n.id, n.location)
        self.neoHandler.add_node(node)

    def way(self, w):
        if w.nodes[0].ref == w.nodes[-1].ref:
            return
        for i in range(len(w.nodes)-1):
            start = GraphNode(w.nodes[i].ref, w.nodes[i].location)
            end = GraphNode(w.nodes[i+1].ref, w.nodes[i+1].location)
            distance = haversine(start, end)
            edge = GraphEdge(start.node_id, end.node_id, distance, 4)
            self.neoHandler.add_edge(edge)
```

```python
class Neo4jHandler():
    def __init__(self):
        self.nodes = []
        self.edges = []

    def add_node(self, node):
        self.nodes.append(node)

    def add_edge(self, edge):
        self.edges.append(edge)
```

```python
url = "sofia_bulgaria.osm.pbf"
neo_handler = Neo4jHandler()
h = FileHandler(neo_handler)

h.apply_file(url, locations=True)
```

```python
data = [{"node_id": n.node_id, "lat": n.lat, "long": n.long} for n in neo_handler.nodes]
stream = iter(data)
batch_size = 25000
```

```python
while True:
    batch = islice(stream, batch_size)
    if batch:
        print("new batch")
        create_nodes(g.auto(), batch, labels={"Node"})
    else:
        break
```

```python
data = [(e.start_node_id, 
         {"distance": e.distance, "rating": e.rating, "cost": e.cost },
         e.end_node_id) for e in neo_handler.edges]
stream = iter(data)
```

```python
batch_size = 1
```

```python
while True:
    batch = islice(stream, batch_size)
    if batch:
        batch_size += 3
        print("new relation")
        merge_relationships(g.auto(), batch,
                            merge_key=("Route"),
                            start_node_key=("Node", "node_id"),
                            end_node_key=("Node", "node_id"))
    else:
        break
```
